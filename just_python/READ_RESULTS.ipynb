```python
TIME = pd.read_parquet(r"output\TIME.parquet")

TIME = TIME[TIME['executor_memory'].notna() & TIME['executor_cores'].notna()]

TIME.loc[:, 'executor_memory'] = TIME['executor_memory'].apply(lambda x: x.replace('g',''))

TIME[['executor_memory', 'executor_cores', 'computation_seconds']] =\
      TIME[['executor_memory', 'executor_cores', 'computation_seconds']].astype(float)

print("Expected average:", TIME['computation_seconds'].mean())

TIME.loc[TIME['Model'] == 'ANN'].pivot_table(index = 'executor_memory', columns = 'executor_cores', values = 'computation_seconds', aggfunc='mean')
```
